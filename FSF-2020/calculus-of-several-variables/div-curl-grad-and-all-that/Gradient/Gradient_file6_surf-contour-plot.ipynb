```python
import numpy as np
print('numpy: '+np.version.full_version)
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.animation as animation
import matplotlib
print('matplotlib: '+matplotlib.__version__)
```

```python
plt.style.use('dark_background')
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111, projection="3d")
ax.xaxis.set_pane_color((1.0, 1.0, 0.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 0.0, 0.0))
ax.zaxis.set_pane_color((1.0, 0.0, 0.0, 0.0))

ax.zaxis._axinfo["grid"]['color'] = (1,1,1,0)
ax.xaxis._axinfo["grid"]['color'] = (1,1,1,0)
ax.yaxis._axinfo["grid"]['color'] = (1,1,1,0)
x, y = np.mgrid[-1:1:30j, -1:1:30j]
z = np.cos(np.pi*x)*np.cos(np.pi*y)
ax.plot_surface(x, y, z+1, cmap="gist_earth", lw=0.5, rstride=1, cstride=1, alpha = 0.8)
ax.contour(x, y, z, 10, cmap="gist_earth", linestyles="solid", offset=-1)
#ax.contour(x, y, z, 3, lw=-10, colors="k", linestyles="solid", offset=-1)
plt.show()
plt.savefig('sc.png')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
x, y = np.mgrid[-1:1:30j, -1:1:30j]
z = 5*np.sin(x)**15 + 20* np.cos(x/y) * np.sin(y-x)
ax.plot_surface(x, y, z, cmap="gist_earth", lw=0.5, rstride=1, cstride=1)
ax.contour(x, y, z, 10, lw=3, cmap="gist_earth", linestyles="solid", offset=-1)
#ax.contour(x, y, z, 10, lw=3, colors="k", linestyles="solid")
plt.show()
```
