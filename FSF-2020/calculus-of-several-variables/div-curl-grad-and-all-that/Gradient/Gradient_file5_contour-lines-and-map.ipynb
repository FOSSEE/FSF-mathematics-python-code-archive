```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import numpy as np
```

```python
def f(x, y):
    return 5*np.sin(x)**15 + 20* np.cos(x/y) * np.sin(y-x)
# np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 8, 20)
y = np.linspace(0, 6, 15)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
a = plt.contourf(X, Y, Z, colors=None, cmap = 'gist_earth')
b = plt.contour(X, Y, Z, colors=None, cmap = 'gist_earth')
```
