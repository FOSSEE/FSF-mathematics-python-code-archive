```python
import numpy as np
import matplotlib.pyplot as plt
Y,X = np.mgrid[-10:10:.01, -10:10:.01]
U = -Y/(X**2 + Y**2)
V = X/(X**2 + Y**2)
plt.streamplot(X,Y, U,V, density=1)
plt.savefig('vortex_streamplot.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
Y,X = np.mgrid[-10:10:.01, -10:10:.01]
U = -Y/(X**2 + Y**2)
V = X/(X**2 + Y**2)
plt.quiver(X,Y, U,V)
plt.show()
```
